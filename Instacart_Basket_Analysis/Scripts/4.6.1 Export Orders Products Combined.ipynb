## Combine orders_products_combined dataframe with products data set. 

```python
# Import Product Dataframe from Prepared data.
df_prods= pd.read_csv(os.path.join(path, 'Data','Prepared Data', 'products_checked.csv'),index_col=0)
```

```python
df_prods.head()
```

```python
# Check Shape
df_prods.shape
```

```python
# Test Merge
pd.merge(orders_products_combined, df_prods, on = ['product_id'], indicator='Match')
```

I used indicator='match' because orders_products_combined already has a _merge column. To use indicator=True in a new merge, I must first drop the existing _merge column.

```python
# Drop Merge Column
orders_products_combined = orders_products_combined.drop(['_merge'], axis=1)
```

```python
orders_products_combined.head()
```

```python
orders_products_combined.shape
```

```python
# Merge Updated Dataframe
df_ords_prods_merge = df_prods.merge(orders_products_combined, on = 'product_id', indicator = True)
```
